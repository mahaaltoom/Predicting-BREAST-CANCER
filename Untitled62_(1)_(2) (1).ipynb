```python
class Breast_Cancer:
  def __init__(self, dataset):
    self.dataset = dataset
  def readDataset(self):
    self.objectDataset = pd.read_csv(self.dataset)
  def  shapeDataset(self):
    print(self.objectDataset.shape)
  def cleanDataset(self):
     self.objectDataset.drop([self.objectDataset.columns[0]], axis= 1, inplace= True)
     self.objectDataset['Class'].loc[self.objectDataset['Class'] == 2] = 0
     self.objectDataset['Class'].loc[self.objectDataset['Class'] == 4] = 1
```

```python
Breast_Cancer_obj = Breast_Cancer('tumor.csv')
```

```python
Breast_Cancer_obj.readDataset()
```

```python
Breast_Cancer_obj.shapeDataset()
```

```python
Breast_Cancer_obj.cleanDataset()
```

```python
Breast_Cancer_obj.objectDataset.head(n=10)
```

```python
x = Breast_Cancer_obj.objectDataset.drop(['Class'],axis=1)
y= Breast_Cancer_obj.objectDataset['Class']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
from sklearn.tree import DecisionTreeClassifier
DecisionTreeClassifier_model = DecisionTreeClassifier(criterion='entropy')
DecisionTreeClassifier_model.fit(x_train,y_train)
```

```python
y_prediction_value = DecisionTreeClassifier_model.predict(x_test)
print("Accuracy Score of decision tree:")
print(accuracy_score(y_test,y_prediction_value))
```

```python
report = classification_report(y_test,y_prediction_value)
print(report)
```

```python
object_ = pd.DataFrame({'Actual values': y_test, 'The values predicted by the model': y_prediction_value})
object_.head(n=10)
```
